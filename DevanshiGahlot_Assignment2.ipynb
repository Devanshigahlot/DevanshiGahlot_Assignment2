# 6.How many Males that have Marital status â€˜Married-civ-spouseâ€™ are earning above 150,000 ?


```python
#First I have filterred the data according to the salary criteria:-
new=df.loc[(df["salary"]>150000)]

#After that I have grouped the data on the basis of marital-status and gender:-
new_1=new.groupby(["marital-status","Gender"])["Gender"].count()
print(new_1)
print(" ")
print(" ")

#After the grouping I have added the indexing to make it more accessible:-
new_1=new.groupby(["marital-status","Gender"], as_index=False)["Gender"].count()
print(new_1)
print(" ")

#Now by the help of the inde3xing I have found the answer to the given question:-
new_1.loc[[4],["marital-status","Gender"]]
```
